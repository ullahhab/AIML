```python
#All imports
import numpy as np
import pandas as pd
#dataset imports to initialize model and test it and migrate to new data
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
#Train and test split auto framework
from sklearn.model_selection import train_test_split
#score and other mmetrics 
from sklearn.metrics import accuracy_score, classification_report
import yfinance as yf
```

```python
#Load iris data for quick model building and testing
iris = load_iris()
x = iris.data
y = (iris.target == 2).astype(int)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
print(x_train)
print(y_train)
```

```python
#Data will be with YF
#We need to categorize data as on x price data for example
#apple data from 22 to 24
data =  yf.download("BTC-USD", start="2022-01-01", end="2025-10-01")

#creating features
data["Return"] = data['Close'].pct_change()
data["MA5"] = data['Close'].rolling(window=5).mean()
data["MA10"] = data['Close'].rolling(window=10).mean()

#print(data)

#drop NAN values don't make an influence on our descision and random noise
data = data.dropna()
print(data)


```

```python
# create target: 1 if tomorrow's close > today's, else 0
data["Target"] = np.where(data['Close'].shift(-1) > data['Close'], 1,0)
x = data[["Return", "MA5", "MA10"]]
y = data["Target"]

#train and test split 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,shuffle=False)
```

```python
#building model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Prediction
print(x_train)
y_pred = model.predict(x_test)
print(y_pred)
```

```python
y_pred = model.predict(x_test)

print("Accuracy: ", accuracy_score(y_test,y_pred))
print("\nClassification report:\n", classification_report(y_test,y_pred))
```

```python
#Next day prediction
latest = x.iloc[-1].values.reshape(1, -1)
next_day_prediction = model.predict(latest)
print("Prediction for next day:", "UP" if next_day_prediction[0] == 1 else "DOWN")
```
